```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import seaborn as sns
```

```python
data_frame = pd.read_csv("transfusion.data")
data_frame
```

```python
data_frame.describe()
```

```python
col1 = "Recency (months)"
col2 = "Frequency (times)"
col3 = "Monetary (c.c. blood)"
col4 = "Time (months)"
col5 = "whether he/she donated blood in March 2007"
```

```python
fig, axs = plt.subplots(3,2,figsize=(15,15))
axs[0][0].plot(data_frame[[col1]],'.',color='r')
axs[0][0].set_title(col1)
axs[0][1].plot(data_frame[[col2]],'.')
axs[0][1].set_title(col2)
axs[1][0].plot(data_frame[[col3]],'.',color = 'y')
axs[1][0].set_title(col3)
axs[1][1].plot(data_frame[[col4]],'.',color = 'g')
axs[1][1].set_title(col4)
axs[2][0].plot(data_frame[[col5]],'.', color = 'c')
axs[2][0].set_title(col5)
fig.delaxes(axs[2][1])
```

```python
data_frame.corr()
```

```python
data_frame = data_frame.iloc[:,[0,1,2,3]]
scatter_matrix(data_frame[data_frame.columns], figsize=(12, 12) , color = 'c' , marker='o' )
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,15))
axs[0][0].boxplot(data_frame[col1])
axs[0][0].set_title(col1)
axs[0][1].boxplot(data_frame[col2])
axs[0][1].set_title(col2)
axs[1][0].boxplot(data_frame[col3])
axs[1][0].set_title(col3)
axs[1][1].boxplot(data_frame[col4])
axs[1][1].set_title(col4)
plt.show()
```

```python
col1 = "Recency (months)"
col2 = "Frequency (times)"
col3 = "Monetary (c.c. blood)"
col4 = "Time (months)"
col5 = "whether he/she donated blood in March 2007"
```
